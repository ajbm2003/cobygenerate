# Generador automático de Razones de Notificación
Genera documentos Word (.docx) a partir del DataFrame con los datos de cada cliente.

```python
import pandas as pd
from docxtpl import DocxTemplate
import os
```

```python
# Leer el DataFrame generado previamente
df = pd.read_excel('NOTIFICACIONESCOACTIVA_OPI_10_02_2026.xlsx')
df
```

```python
df_2= pd.read_csv('datos_capturados.csv')

# Mapear meses en español a inglés para poder parsear las fechas
meses_es_en = {
    'ene': 'Jan', 'feb': 'Feb', 'mar': 'Mar', 'abr': 'Apr',
    'may': 'May', 'jun': 'Jun', 'jul': 'Jul', 'ago': 'Aug',
    'sept': 'Sep', 'sep': 'Sep', 'oct': 'Oct', 'nov': 'Nov', 'dic': 'Dec'
}

fecha_en = df_2['Fecha Envío CPanel'].copy()
for es, en in meses_es_en.items():
    fecha_en = fecha_en.str.replace(f' {es} ', f' {en} ', case=False, regex=False)

df_2['Fecha Envío CPanel_dt'] = pd.to_datetime(fecha_en, format='%d %b %Y %H:%M:%S')

# Filtrar por remitente y fechas 11-12 feb 2026
df_2 = df_2[
    (df_2['Remitente'] == 'cobranzaypatrocinio@cobypat.com') &
    (df_2['Fecha Envío CPanel_dt'].dt.date.isin([
        pd.Timestamp('2026-02-11').date(),
        pd.Timestamp('2026-02-12').date()
    ]))
].drop(columns=['Fecha Envío CPanel_dt'])

df_2
```

```python
# Diccionario: Destinatario → [lista de fechas de notificación]
# Excluyendo cobranzaypatrocinio@cobypat.com
dic_fechas = (
    df_2[df_2['Destinatario'] != 'cobranzaypatrocinio@cobypat.com']
    .groupby('Destinatario')['Fecha Envío CPanel']
    .apply(list)
    .to_dict()
)

print(f"{len(dic_fechas)} destinatarios únicos")
dic_fechas
```

```python
# Crear diccionario: NUMERO_TITULO → [lista de emails asociados]
new_dic = df.groupby('NUMERO_TITULO')['Email'].apply(list).to_dict()

# Crear dic_fechas normalizado (llaves en minúsculas)
dic_fechas_lower = {k.lower().strip(): v for k, v in dic_fechas.items()}

# Asignar FECHA_NOTIFICACION buscando cada email en dic_fechas
def obtener_fechas(email):
    fechas = dic_fechas_lower.get(str(email).lower().strip(), [])
    return ', '.join(fechas) if fechas else ''

# Forzar la actualización (eliminar columna si ya existe para recalcular)
if 'FECHA_NOTIFICACION' in df.columns:
    df.drop(columns=['FECHA_NOTIFICACION'], inplace=True)

df['FECHA_NOTIFICACION'] = df['Email'].apply(obtener_fechas)

print(f"{len(new_dic)} NUMERO_TITULO únicos")
print(f"Filas con fechas asignadas: {(df['FECHA_NOTIFICACION'] != '').sum()} de {len(df)}")
df[['NUMERO_TITULO', 'Email', 'FECHA_NOTIFICACION']].head(10)
```

```python
# ============================================================
# Generar documentos Word a partir del diccionario new_dic
# ============================================================

from docx import Document

PLANTILLA = 'RAZÓN DE NOTIFICACIÓN 1.docx'
output_dir = 'razones_notificacion'
os.makedirs(output_dir, exist_ok=True)

# Mapeo de abreviaciones a nombres completos en español
meses_completos = {
    'ene': 'Enero', 'feb': 'Febrero', 'mar': 'Marzo', 'abr': 'Abril',
    'may': 'Mayo', 'jun': 'Junio', 'jul': 'Julio', 'ago': 'Agosto',
    'sept': 'Septiembre', 'sep': 'Septiembre', 'oct': 'Octubre',
    'nov': 'Noviembre', 'dic': 'Diciembre'
}

def expandir_mes(texto_fecha):
    """Reemplaza abreviación del mes por nombre completo en español.
    Ej: '11 feb 2026 10:30:45' → '11 de febrero de 2026 10:30:45'"""
    texto = str(texto_fecha).strip()
    for abrev, completo in meses_completos.items():
        if f' {abrev} ' in texto.lower():
            # Reemplazar la abreviación conservando la estructura
            partes = texto.split()
            for i, p in enumerate(partes):
                if p.lower() == abrev:
                    partes[i] = f'de {completo} de'
                    break
            return ' '.join(partes)
    return texto

def formatear_fechas_notificacion(fecha_notificacion):
    """Divide FECHA_NOTIFICACION por ',' y devuelve texto 'fecha1 y fecha2'
    con meses en nombre completo."""
    if not fecha_notificacion or pd.isna(fecha_notificacion) or str(fecha_notificacion).strip() == '':
        return ''
    partes = [expandir_mes(f.strip()) for f in str(fecha_notificacion).split(',') if f.strip()]
    if len(partes) == 0:
        return ''
    elif len(partes) == 1:
        return partes[0]
    else:
        return ' y '.join(partes)

def reemplazar_en_parrafo(parrafo, reemplazos):
    """Reemplaza marcadores [CAMPO] en un párrafo conservando el formato."""
    for marcador, valor in reemplazos.items():
        if marcador not in parrafo.text:
            continue
        for run in parrafo.runs:
            if marcador in run.text:
                run.text = run.text.replace(marcador, str(valor))
        if marcador in ''.join(r.text for r in parrafo.runs):
            texto_completo = parrafo.text
            if marcador in texto_completo:
                texto_nuevo = texto_completo.replace(marcador, str(valor))
                if parrafo.runs:
                    for j in range(len(parrafo.runs) - 1, 0, -1):
                        parrafo.runs[j].text = ''
                    parrafo.runs[0].text = texto_nuevo

def reemplazar_en_documento(doc, reemplazos):
    """Reemplaza marcadores en todo el documento."""
    for parrafo in doc.paragraphs:
        reemplazar_en_parrafo(parrafo, reemplazos)
    for tabla in doc.tables:
        for fila in tabla.rows:
            for celda in fila.cells:
                for parrafo in celda.paragraphs:
                    reemplazar_en_parrafo(parrafo, reemplazos)
    for section in doc.sections:
        for parrafo in section.header.paragraphs:
            reemplazar_en_parrafo(parrafo, reemplazos)
        for parrafo in section.footer.paragraphs:
            reemplazar_en_parrafo(parrafo, reemplazos)

# ============================================================

if not os.path.exists(PLANTILLA):
    print(f"⚠️  No se encontró la plantilla '{PLANTILLA}'.")
else:
    count = 0
    for numero_titulo, emails in new_dic.items():
        datos_cliente = df[df['NUMERO_TITULO'] == numero_titulo].iloc[0]
        correos_unidos = ', '.join(emails)
        
        # Formatear fechas: "11 feb 2026 10:30:45, 12 feb 2026 14:20:00"
        # → "11 de febrero de 2026 10:30:45 y 12 de febrero de 2026 14:20:00"
        fechas_texto = formatear_fechas_notificacion(datos_cliente.get('FECHA_NOTIFICACION', ''))
        
        doc = Document(PLANTILLA)
        
        reemplazos = {
            'TITULO_CREDITO': str(datos_cliente['CUENTA_CONTRATO']),
            'NOMBRE_CLIENTE': str(datos_cliente['NOMBRE_CLIENTE']).title(),
            'CEDULA_CLIENTE': str(numero_titulo),
            'CORREO': correos_unidos,
            'FECHAS': fechas_texto,
        }
        
        reemplazar_en_documento(doc, reemplazos)
        
        filename = f"{output_dir}/Razon_{datos_cliente['CUENTA_CONTRATO']}_{numero_titulo}.docx"
        doc.save(filename)
        count += 1
        print(f"Generado: {filename}  →  Fechas: {fechas_texto}")
    
    print(f"\nSe generaron {count} documentos (1 por cédula) en '{output_dir}'")
```

```python
import requests

def detectar_genero(nombre_completo):
    """Detecta el género con doble criterio:
    1) Mayoría: si 3+ palabras son male o female, se decide por mayoría.
    2) Empate: si hay igual cantidad de male y female, se decide por promedio de probabilidad."""
    
    palabras = nombre_completo.strip().split()
    probs_male = []
    probs_female = []
    
    for palabra in palabras:
        palabra_title = palabra.title()
        try:
            resp = requests.get("https://api.genderize.io", 
                              params={"name": palabra_title, "country_id": "EC"})
            data = resp.json()
            g = data.get("gender")
            p = data.get("probability", 0)
            print(f"  '{palabra_title}' → género: {g}, probabilidad: {p}")
            
            if g == "male":
                probs_male.append(p)
            elif g == "female":
                probs_female.append(p)
        except Exception as e:
            print(f"  Error con '{palabra_title}': {e}")
    
    n_male = len(probs_male)
    n_female = len(probs_female)
    
    print(f"\n  Conteo male: {n_male} | Conteo female: {n_female}")
    
    # Criterio 1: Mayoría (si uno tiene más coincidencias que el otro)
    if n_male != n_female:
        if n_male > n_female:
            print(f"  → Decisión por MAYORÍA: male ({n_male} vs {n_female})")
            return "male", sum(probs_male) / n_male
        else:
            print(f"  → Decisión por MAYORÍA: female ({n_female} vs {n_male})")
            return "female", sum(probs_female) / n_female
    
    # Criterio 2: Empate → decidir por promedio de probabilidad
    if n_male > 0 and n_female > 0:
        prom_male = sum(probs_male) / n_male
        prom_female = sum(probs_female) / n_female
        print(f"  Empate ({n_male} vs {n_female}) → Promedio male: {prom_male:.4f} | Promedio female: {prom_female:.4f}")
        if prom_female >= prom_male:
            print(f"  → Decisión por PROBABILIDAD: female")
            return "female", prom_female
        else:
            print(f"  → Decisión por PROBABILIDAD: male")
            return "male", prom_male
    
    # Si no hubo coincidencias
    return None, 0

# --- Pruebas ---
nombres_prueba = [
    "QUIROZ MARQUEZ MARTHA INES"
]

for nombre in nombres_prueba:
    print(f"\n{'='*50}")
    print(f"Nombre completo: {nombre}")
    genero, prob = detectar_genero(nombre)
    
    if genero == "female":
        termino = "Coactivada"
    else:
        termino = "Coactivado"
    
    print(f"\n→ Resultado: {genero} (prob: {prob:.4f}) → {termino}")
```
